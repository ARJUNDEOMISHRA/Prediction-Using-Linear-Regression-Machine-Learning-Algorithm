## model 3: GRu 
Another popular and effetive Rnn component is the GRu or gated recurrent unit.
The Gru cell has similar feature to an lstm cell but has less parameters

```python
import tensorflow as tf
from tensorflow.keras import layers

# Example setup (assuming you already have a text_vectorizer and embedding layer)
inputs = layers.Input(shape=(1,), dtype=tf.string)

x = text_vectorizer(inputs)
x = embedding(x)

# Stacked RNN architecture
x = layers.GRU(64, return_sequences=True)(x)
x = layers.LSTM(64, return_sequences=True)(x)   # must output sequences
x = layers.GRU(64)(x)                           # final recurrent layer (no sequences)
x = layers.Dense(64, activation="relu")(x)
outputs = layers.Dense(1, activation="sigmoid")(x)

model_3 = tf.keras.Model(inputs, outputs, name="model_3_GRU")

model_3.summary()
```

```python
import tensorflow as tf
import tensorflow_hub as hub
from tensorflow.keras import layers
from tensorflow.keras.losses import BinaryCrossentropy

# Wrap USE inside a custom Keras layer
class USEWrapper(tf.keras.layers.Layer):
    def __init__(self, model_url):
        super().__init__()
        self.model = hub.load(model_url)  # load as TF function, not KerasLayer

    def call(self, inputs):
        return self.model(inputs)  # call TF Hub model inside a Keras Layer

# Instantiate wrapped USE layer
use_layer = USEWrapper("https://tfhub.dev/google/universal-sentence-encoder/4")

# Build Functional model
inputs = tf.keras.Input(shape=(), dtype=tf.string, name="text_input")
x = use_layer(inputs)
x = layers.Dense(64, activation="relu")(x)
outputs = layers.Dense(1, activation="sigmoid")(x)

model_6 = tf.keras.Model(inputs, outputs, name="model_6_USE")

# Compile
model_6.compile(
    loss=BinaryCrossentropy(),
    optimizer=tf.keras.optimizers.Adam(),
    metrics=["accuracy"]
)

# Summary
model_6.summary()

# Test
sample_text = tf.constant(["TensorFlow Hub makes text embeddings easy"])
pred = model_6.predict(sample_text)
print(pred)
```

```python
import tensorflow as tf
import tensorflow_hub as hub
from tensorflow.keras.layers import Layer

# Custom USE wrapper with from_config
class USEWrapper(Layer):
    def __init__(self, model_url="https://tfhub.dev/google/universal-sentence-encoder/4", trainable=False, **kwargs):
        super().__init__(**kwargs)
        self.model_url = model_url
        self.trainable_flag = trainable
        self.use = hub.KerasLayer(self.model_url, trainable=self.trainable_flag, dtype=tf.string)

    def call(self, inputs):
        return self.use(inputs)

    def get_config(self):
        config = super().get_config()
        config.update({
            "model_url": self.model_url,
            "trainable": self.trainable_flag
        })
        return config

    @classmethod
    def from_config(cls, config):
        # Keras will pass the config dict here during deserialization
        return cls(**config)

# Now load the model
loaded_model_6 = tf.keras.models.load_model(
    "model_6.h5",
    custom_objects={"USEWrapper": USEWrapper}
)

loaded_model_6.summary()
```

```python
# Let's make a function to measure the time of prediction
import time
def pred_timer(model, samples):
    """
    Times how long a model takes to make predictions on samples.
    """
    start_time = time.perf_counter() # get start time
    model.predict(samples) # make predictions
    end_time = time.perf_counter() # get finish time
    total_time = end_time - start_time # calcluate how long predictons took to make
    time_per_pred = total_time/len(samples)
    return total_time, time_per_pred
```

```python
import time
import tensorflow as tf

def pred_timer(model, samples):
    """
    Times how long a TFSMLayer model takes to make predictions on samples.

    Args:
        model: TFSMLayer object
        samples: list of strings (sentences)

    Returns:
        total_time: float, total time in seconds
        time_per_pred: float, average time per sample in seconds
    """
    # Convert list of strings to tensor
    samples_tensor = tf.constant(samples)

    start_time = time.perf_counter()
    output = model(samples_tensor)  # Call TFSMLayer directly
    end_time = time.perf_counter()

    total_time = end_time - start_time
    time_per_pred = total_time / len(samples)

    return total_time, time_per_pred
```
